```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns;
```

```python
#show all columns
df = pd.read_excel('KOV785TZ111.xls')
pd.options.display.max_columns = None
df.head()
```

```python
#select needed columns
df_w = df[['NR','TAR_LEDA','SEX','MMUZA','SACH_MISHKL','CD_1','CHU1','TZI1']]
df_w
```

```python
#drop rows where MMUZA is zero or TZI1 is zero
df_w = df_w.drop(df_w[(df_w.MMUZA == 0) | (df_w.TZI1 == 0)].index)
df_w
```

```python
#check corr between MMUZA & TZI1
df_w[['TZI1','MMUZA']].corr()
#plt.matshow(df_w_cor)
```

```python
df_w.CD_1.nunique()
df_w.CHU1.nunique()
```

```python
d = {}
for i in df_w.CD_1.unique():
    corr_i = df_w[df_w.CD_1 == i].TZI1.corr(df_w[df_w.CD_1 == i].MMUZA)
    if abs(corr_i) >= 0.7:
        #print(df_w.CHU1[df_w.CD_1 == i][:1].values[0] ,round(corr_i,2))
        d[df_w.CHU1[df_w.CD_1 == i][:1].values[0]] = round(corr_i,2)

s = [(k, d[k]) for k in sorted(d, key=d.get, reverse=True)]
s

```

```python
x, y = zip(*s) # unpack a list of pairs into two tuples
data = pd.Series(y, index=x)
ax = data.plot(title="Avg in 1st Degree VS Current Avg ",kind='bar',figsize=(7,5))
_=ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
#ax.set_xticks(np.arange(len(x)))
#data.plot.bar()
```
