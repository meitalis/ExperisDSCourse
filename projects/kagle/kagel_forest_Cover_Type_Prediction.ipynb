### https://www.kaggle.com/c/forest-cover-type-prediction

The training set (15120 observations) contains both features and the Cover_Type. The test set contains only the features. You must predict the Cover_Type for every row in the test set (565892 observations).

```python
import pandas as pd 
from sklearn import datasets
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV


from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection  import cross_val_score
```

```python
df = pd.read_csv('forest_Cover_Type/train.csv')
```

# train

```python
X = df.drop('Cover_Type',axis=1)
y = df.Cover_Type

X_train,X_test,y_train ,y_test = train_test_split(X,y,random_state = 42,train_size = 0.8,test_size=0.2)
X_train,X_test ,y_train ,y_test

dtc_cls = DecisionTreeClassifier()

param_grid = {'max_leaf_nodes':np.arange(2,50,2),
             'criterion' :['gini','entropy'],
             'splitter' : ['best','random']}
             


grid_search = GridSearchCV(dtc_cls,param_grid,cv=10)
grid_search.fit(X_train,y_train)

print('best_params_',grid_search.best_score_,grid_search.best_params_)

results = grid_search.cv_results_
for mean_score,params in zip(results['mean_test_score'],results['params']):
    print(mean_score,params)

# best_params_ 0.7014715608465608 {'criterion': 'gini', 'max_leaf_nodes': 48, 'splitter': 'best'}

```

```python
dtc_cls_best = DecisionTreeClassifier(criterion='gini', max_leaf_nodes=48, splitter='best')
scores = cross_val_score(dtc_cls_best,X_test,y_test,cv =10)
scores,scores.mean(),scores.std()
#mean 0.6898819319990703,
```

```python
X = df.drop('Cover_Type',axis=1)
y = df.Cover_Type

X_train,X_test,y_train ,y_test = train_test_split(X,y,random_state = 42,train_size = 0.8,test_size=0.2)
X_train,X_test ,y_train ,y_test
    
    
clf = Pipeline([('scale',StandardScaler()),
               ('cls',SVC())])

#'poly', 'rbf', 'sigmoid', 'precomputed'
param_grid = {'cls__C':np.arange(2,10,2),
             'cls__kernel' :['linear','poly']}

grid_search = GridSearchCV(clf,param_grid,cv=10)
grid_search.fit(X_train,y_train)

print('best_params_',grid_search.best_score_,grid_search.best_params_)

results = grid_search.cv_results_
for mean_score,params in zip(results['mean_test_score'],results['params']):
    print(mean_score,params)
```

```python
#best_params_ 0.7234623015873016 {'cls__C': 2, 'cls__kernel': 'linear'}
#best_params_ 0.7658730158730159 {'cls__C': 8, 'cls__kernel': 'poly'}

clf = Pipeline([('scale',StandardScaler()),
               ('cls',SVC(C=8,kernel='poly'))])

scores = cross_val_score(clf,X_test,y_test,cv =10)
scores,scores.mean(),scores.std()
```

```python
#RandomForest - try
X = df.drop('Cover_Type',axis=1)
y = df.Cover_Type

X_train,X_test,y_train ,y_test = train_test_split(X,y,random_state = 42,train_size = 0.8,test_size=0.2)
X_train,X_test ,y_train ,y_test

    
clf = Pipeline([('scale',StandardScaler()),
               ('cls',RandomForestClassifier())])

#'poly', 'rbf', 'sigmoid', 'precomputed'
param_grid = {'cls__n_estimators':np.arange(2,10,2),
             'cls__criterion' :['gini','entropy'],
             'cls__max_depth':np.arange(2,10,2)}

grid_search = GridSearchCV(clf,param_grid,cv=10)
grid_search.fit(X_train,y_train)

print('best_params_',grid_search.best_score_,grid_search.best_params_)

results = grid_search.cv_results_
for mean_score,params in zip(results['mean_test_score'],results['params']):
    print(mean_score,params)
    
#best_params_ 0.7353672524933523 {'cls__criterion': 'entropy', 'cls__max_depth': 8, 'cls__n_estimators': 8}
```
