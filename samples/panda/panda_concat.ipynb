```python
import pandas as pd
import numpy as np  
```

```python
s1 = pd.Series([0,1],index=['a','b'])
s2 = pd.Series([2,3,4],index=['c','d','e'])
s3 = pd.Series([5,6],index=['f','g'])

#ex1
pd.concat([s1,s2,s3])
#ex2
pd.concat([s1,s2,s3],axis=1)
#ex3
s4 = pd.concat([s1 * 5,s3])
s4
#ex4
pd.concat([s1,s4],axis = 1)
#ex5
print(s1,s4)
pd.concat([s1,s4],axis=1,join='inner')

pd.concat([s1,s4],axis=1,join='outer')

print(s1,s4)
pd.concat([s1,s4],axis=1,join_axes=[['a','c','b','e']])

print(s1,'\n',s3)
pd.concat([s1,s1,s3],keys=['ones','two','three'])

print(s1,'\n',s2,'\n',s3)
pd.concat([s1,s2,s3],axis = 1,keys=['ones','two','three'])
```

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})


df3 = pd.merge(df1,df2)
df3_2 = pd.merge(df1,df2,on='employee')

display(df1,df2,df3,df3_2)
```

```python
#concat not good in this case we need merge
pd.concat([df1,df2]) 
pd.concat([df1,df2],axis=1) 
```

```python
#many to one
df4 = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR'],
                    'supervisor': ['Carly', 'Guido', 'Steve']})
print(df3,'\n',df4)

display(pd.merge(df3,df4))
```

```python
#many to many
df5 = pd.DataFrame({'group': ['Accounting', 'Accounting',
                              'Engineering', 'Engineering', 'HR', 'HR'],
                    'skills': ['math', 'spreadsheets', 'coding', 'linux',
                               'spreadsheets', 'organization']})

display(df1,df5,pd.merge(df1,df5))
```

```python
df6 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})
df7 = pd.merge(df1,df6,left_on='employee',right_on='name')
display(df1,df6,df7)
```
