```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sales Records.csv')
```

```python
#explore data set
display(df.shape,df.tail(),df.head(),df.columns,df.info(),df.describe().T)
display(df.isna().sum(),df.nunique())
display(df.isna().sum().sum())
#display(df.day.unique(),df.time.unique(),df['size'].unique())
```

```python
df.head()
```

```python
#column names -> replace " " with "_" and lower case
df.columns = df.columns.str.replace(" ", "_").str.lower()
df.head()
```

```python
#what is the most profitable item
df['profit_perc_per_unit'] = 100 * (df.unit_price - df.unit_cost) / df.unit_cost
display(df.head(5))
df.groupby('item_type')['profit_perc_per_unit'].mean()
df.groupby('item_type')['profit_perc_per_unit'].mean().sort_values(ascending=False).plot.bar();
```

```python
#in which region buying more online (units_sold)
#display(df[df.sales_channel=='Online'])
df[df.sales_channel=='Online'].groupby('region')['units_sold'].sum().sort_values(ascending=False).plot.bar();
```
