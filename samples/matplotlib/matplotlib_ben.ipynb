```python
s = np.linspace(0,1,100)
k = 0.04
v_max = 0.1 
v = (v_max * s) / (k + s)
fig = plt.figure()

plt.figure(facecolor="white")
plt.title(' Michaelis-Menten curve for an enzyme reaction')
plt.grid(axis='y')
plt.ylabel('reaction rate [v]')
plt.xlabel('substrate [s]')

plt.plot(s,v);
```

```python
years = np.array([1972,1974,1978,1982,1985,1989,1993,1997,1999,2000,2003,2004,2007,2008,2012])
observed = np.array([0.0025,0.005,0.029,0.12,0.275,1.18,3.1,7.5,24,42,220,592,1720,2046,3100])
observed = observed * (10**6)
log_observed = np.log10(observed)

predicted = np.log10(observed[0]) + ((years-1972)/2)*np.log10(2)

```

```python
fig,axes = plt.subplots(1)
axes.plot(years,log_observed,'r*',label='observed')
axes.plot(years,predicted,'--k',label='predicted')

axes.legend();

axes.set_title("Moore's Law");
axes.set_xlabel("year")
axes.set_ylabel("log")
```
