```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = load_boston()
```

```python

print(type(data.target))
# data['target'] - 1D numpy.ndarray

print(type(data.data))
# data['target'] - 2D numpy.ndarray

print(type(data.feature_names))
# data['target'] - 1D numpy.ndarray

print(type(data.DESCR))
# data['target'] - str
```

```python
print(data.DESCR)
print(data.feature_names)
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df['target'] = pd.Series(data.target)
```

```python
df.describe()
```

```python
pd.isnull(df).any()
```

```python
pd.isnull(df).sum()
```

```python
all_corr = df.corr()
all_corr
```

```python
#how each input attribute is able to predict the target
# assume target attr is the last, then remove corr with itself
corr_attr_target = all_corr.iloc[-1][:-1]
corr_attr_target = corr_attr_target[abs(corr_attr_target).argsort()[::-1]]
corr_attr_target              
# - RM       average number of rooms per dwelling
# - PTRATIO  pupil-teacher ratio by town
# - LSTAT    % lower status of the population
```

```python
#correlations between attribute pairs
attrs_corr = all_corr.iloc[:-1,:-1] # all except target
threshold = 0.5
attrs_corr = (attrs_corr[abs(attrs_corr) > 0.5][attrs_corr != 1.0]).unstack().dropna().to_dict()

unique_corrs = pd.DataFrame(list(set([(tuple(sorted(key)), attrs_corr[key]) \
    for key in attrs_corr])), columns=['attributes', 'correlation'])


unique_corrs = unique_corrs.iloc[abs(unique_corrs['correlation']).argsort()[::-1]]
unique_corrs

#(RAD, TAX)	0.910228
# - RAD      index of accessibility to radial highways
# - TAX      full-value property-tax rate per $10,000
#מדד הנגישות לכבישים רדיאליים
#שיעור מס רכוש בשווי מלא לכל 10,000 דולר
```
