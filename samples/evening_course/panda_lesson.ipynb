```python
df4['targets'] = data.target
```

```python
#remove 3 columns
df4.drop(['mean area', 'mean concave points', 'mean symmetry'], axis=1, inplace=True)
```

```python
df4
```

```python
df5 = pd.concat([df4.loc[:,df4.columns[1:5:1]], df4.loc[:,df4.columns[15:20:1]]], axis=1)
```

```python

df5
```

```python
df5.tail()
```

```python
#show true/false per column name
df5['mean texture']>25

#show the data frame in lines with true - remove lines with false
df5[df5['mean texture']>25]

# we can filter according to complex condition of and/or but we must use panda specific operators: '& and '|'
df5[(df5['mean texture']>25) & (df5['mean perimeter']<79)]
```

```python
df6 = df5[(df5['symmetry error']>0.02) & (df5['symmetry error']<0.025)]
df6
```
